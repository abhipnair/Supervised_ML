# Hand Written Image Classification with Confusion Matrix.

```python
from sklearn.datasets import load_digits
```

```python
# Load the digits dataset
digits = load_digits()
```

```python
# looking into data. 
print(digits.data[0]) # type: ignore
```

```python
import matplotlib.pyplot as plt

# Code for showing images. 
plt.gray()
# plt.matshow(digits.images[0])
for i in range(3):
    plt.matshow(digits.images[i]) # type: ignore
```

```python
# using function from sklearn module to import a function train_test_slipt
# the function when passed an df 

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2) #type:ignore
```

```python
from sklearn.linear_model import LogisticRegression
handWritten_num_model = LogisticRegression(solver='newton-cg')
```

```python
handWritten_num_model.fit(x_train, y_train)
```

```python
plt.matshow(digits.images[1477])  # type: ignore
```

```python
handWritten_num_model.predict(digits.data[[1477]])   # type: ignore
# len(x_train)S
```

```python
handWritten_num_model.score(x_test, y_test)
```

# For getting where the model outperformes

**we can use confusion matrix**

```python
from sklearn.metrics import confusion_matrix
y_predicted = handWritten_num_model.predict(x_test)

cm = confusion_matrix(y_test, y_predicted) 

cm
```

```python
# ploting the confusion matrix. 

plt.figure(figsize=(10, 7))
plt.imshow(cm, cmap='viridis', interpolation='nearest')

# Add annotations
for i in range(len(cm)):
    for j in range(len(cm[i])):
        # Display the values from 'cm' directly as annotations
        plt.text(j, i, str(cm[i, j]), ha='center', va='center', color='white')

plt.xlabel("Predicted")
plt.ylabel("Truth")

# Add a colorbar
cbar = plt.colorbar()
cbar.set_label("Annotation")

plt.show()
```

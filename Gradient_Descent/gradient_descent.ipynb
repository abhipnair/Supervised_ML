# Gradient Descent. 

**Gradient descent is a fundamental optimization algorithm used in machine learning and numerical optimization. It is employed to find the minimum (or maximum) of a function, typically referred to as the "cost function" or "loss function." Gradient descent is widely used in training machine learning models, particularly for tasks like linear regression, logistic regression, neural network training, and more.**

```python
import numpy as np
import pandas as pd
```

```python
m_curr=b_curr=0 # it refers to m and b ie. slope and intercept
learning_rate = 0.001 # refers to the step taken by the algorithm in order to find the m and b. It is tunable. 
iterations = 1000 # it is used for the purposes such as parameter updates, learning rate optimisation etc. 


# the function which calculates the m and n based on the x and y inputs. 
def gradient_descent(x, y, n):
    m_curr=b_curr=0

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr


        cost_fun = 1/n * sum([val**2 for val in ( y- y_predicted)]) # cost function is mean squared error
        m_deri = -(2/n) * sum(x*(y - y_predicted)) # partial derivative of mean squared error with respect to m. 
        b_deri = -(2/n) * sum(( y- y_predicted)) # partial derivative of mean squared error with respect to b. 

        m_curr = m_curr - learning_rate * m_deri
        b_curr = b_curr - learning_rate * b_deri
        print("m {}, b {}, cost {}, iteration {}".format(m_curr, b_curr, cost_fun, i))
```

```python
# Reading th data
data = pd.read_csv("test_scores.csv")
n = len(data.math)
data
# print(n)
```

```python
# Scaling the data (you can use other scaling methods if needed)
x_scaled = (data.math - data.math.mean()) / data.math.std()
y_scaled = (data.cs - data.cs.mean()) / data.cs.std()

gradient_descent(x_scaled, y_scaled, n)

# you can modify this code by giving testing data to the model. In this we had only found the nessasities which we want to compute the preditions. 
```

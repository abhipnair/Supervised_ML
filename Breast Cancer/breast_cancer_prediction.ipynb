# Breast Cancer Prediction Model. 

```python
from sklearn.datasets import load_breast_cancer # importing the dataset
```

```python
breast_cancer = load_breast_cancer() # loading the object. 
hi = "love"
dir(breast_cancer) # looking into the object contents. 
```

**Data Preprocessing**

```python
print(breast_cancer.DESCR) # looking into the description of the dataset. 
```

```python
print(breast_cancer.target) # displaying target. 
```

```python
import pandas as pd
breast_cancer_data = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names) # creating a table or dataframe
breast_cancer_data['target'] = breast_cancer.target # adding target (y)
breast_cancer_data['type'] = breast_cancer_data.target.apply(lambda x: breast_cancer.target_names[x]) # enhancing the data with more information
breast_cancer_data
```

```python
# spliting data for training
X_data = breast_cancer_data.drop(['target', "type"], axis='columns')
y_data = breast_cancer_data['target'] 
X_data.head(0)
```

**Creation and training of the model**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data) # further spliting a data
```

```python
from sklearn.svm import SVC
model = SVC(kernel="rbf", C=3) # creating the Support Vector Machine model obj with kernal = rbf and Regularization as 3
```

```python
model.fit(X_train, y_train) # training the model with avilable data. 
model.fit_status_ # if successful or not
```

```python
model.score(X_test, y_test) # predicting the score. 
```
